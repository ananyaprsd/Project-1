```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/PlayTennis.csv")
import seaborn as sns
import sklearn as sk
data.info()
data.columns
data.head(15)
x_train=pd.get_dummies(data.drop(['Play Tennis'],axis=1))
y_train=data['Play Tennis']
print(x_train)
print(y_train)
from sklearn.naive_bayes import GaussianNB
a=pd.get_dummies(data['Play Tennis'],drop_first=True)
print(a)
model=GaussianNB()
model.fit(x_train,a)
```

```python
predicted=model.predict([[1,1,0,1,1,1,1,1,1,1]])
print(predicted)
predicted=model.predict([[0,1,0,1,1,1,1,1,1,1]])
print(predicted)
```
