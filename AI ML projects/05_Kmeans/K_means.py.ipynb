```python
import matplotlib.pyplot as plt
```

```python
image=plt.imread('/content/butterfly.avif')
plt.imshow(image)
```

```python
image.shape
```

```python
from sklearn.cluster import KMeans
```

```python
X=image.reshape(-1,3)
print(X.shape)
```

```python
kmeans=KMeans(n_clusters=5,n_init=10)
kmeans.fit(X)
```

```python
segmented_img=kmeans.cluster_centers_[kmeans.labels_]
segmented_img1=segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img1/255)
```
